# Analisis Bivariat

```python
data= pd.DataFrame(data)
gender_interest_count = data.groupby(['Interest','Gender']).size().unstack().fillna(0)
plt.figure(figsize=(10,6))
gender_interest_count.plot(kind='bar', stacked=False, color=('salmon','skyblue'))

plt.title('Interest Distribution Based On Gender')
plt.xlabel('Interest')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.legend(title='Gender')
plt.tight_layout()
plt.show()
```

```python
data= pd.DataFrame(data)
gender_interest_count = data.groupby(['Personality','Gender']).size().unstack().fillna(0)
plt.figure(figsize=(10,6))
gender_interest_count.plot(kind='bar', stacked=False, color=('salmon','skyblue'))

plt.title('Personality Distribution Based On Gender')
plt.xlabel('Personality')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.legend(title='Gender')
plt.tight_layout()
plt.show()
```

```python
data= pd.DataFrame(data)
gender_interest_count = data.groupby(['Personality','Interest']).size().unstack().fillna(0)
plt.figure(figsize=(12,10))
gender_interest_count.plot(kind='bar', stacked=False)

plt.title('Personality Distribution Based On Interest')
plt.xlabel('Personality')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.legend(title='Interest')
plt.tight_layout()
plt.show()
```
