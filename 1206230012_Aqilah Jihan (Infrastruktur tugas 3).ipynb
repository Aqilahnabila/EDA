## Korelasi

```python
k=pearsonr(data['Age'],data['Introversion Score'])
print('Berdasarkan',k,'dapat disimpulkan bahwa korelasi antara Age dengan Introversion Score lemah dan tidak signifikan')
```

```python
k2=pearsonr(data['Age'],data['Sensing Score'])
print('Berdasarkan',k2,'dapat disimpulkan bahwa korelasi antara Age dengan Sensing Score lemah dan tidak signifikan')
```

```python
k3= pearsonr(data['Age'],data['Thinking Score'])
print('Berdasarkan',k3,'dapat disimpulkan bahwa korelasi antara Age dengan Thinking Score lemah dan tidak signifikan')
```

```python
k4= pearsonr(data['Age'],data['Judging Score'])
print('Berdasarkan',k4,'dapat disimpulkan bahwa korelasi antara Age dengan Judging Score lemah dan tidak signifikan')
```

```python
data= pd.DataFrame(data)
gender_interest_count = data.groupby(['Interest','Gender']).size().unstack().fillna(0) # unstack() untuk mengubah hasil groupby() mnjdi dataframe yang mudah diolah
plt.figure(figsize=(10,6))
gender_interest_count.plot(kind='bar', stacked=False, color=('salmon','skyblue')) # stacked=false untuk membuat bar chart biasa dengan setiap kategorinya digambarkan sebagai bar terpisah

plt.title('Interest Distribution Based On Gender')
plt.xlabel('Interest')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.legend(title='Gender')
plt.tight_layout()
plt.show()
```

#### Berdasarkan grafik tersebut dapat disimpulkan bahwa perbandingan gender dengan interest banyak yang tidak diketahui. Namun, Diantara lainnya (selain unknown) Arts memiliki frekuensi tertinggi yang digemari baik untuk male or female

```python
data= pd.DataFrame(data)
gender_interest_count = data.groupby(['Personality','Gender']).size().unstack().fillna(0)
plt.figure(figsize=(10,6))
gender_interest_count.plot(kind='bar', stacked=False, color=('salmon','skyblue'))

plt.title('Personality Distribution Based On Gender')
plt.xlabel('Personality')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.legend(title='Gender')
plt.tight_layout()
plt.show()
```

#### Berdasarkan grafik diatas dapat disimpulkan bahwa personality ENFP banyak dimiliki oleh male maupun female

```python
data= pd.DataFrame(data)
gender_interest_count = data.groupby(['Personality','Interest']).size().unstack().fillna(0)
plt.figure(figsize=(12,10))
gender_interest_count.plot(kind='bar', stacked=False)

plt.title('Personality Distribution Based On Interest')
plt.xlabel('Personality')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.legend(title='Interest')
plt.tight_layout()
plt.show()
```

#### Berdasarkan grafik diatas dapat disimpulkan bahwa setiap interest yang memiliki frekuensi tertinggi memiliki personality ENFP
